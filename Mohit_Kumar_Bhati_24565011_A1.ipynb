## *Introduction to the dataset*  

Importing Necessary Libraries and Loading the Dataset

```python
import pandas as pd

df = pd.read_csv('Spotify Most Streamed Songs.csv')
df.head()
```

Describe the dataset

```python
df.info()
df.describe()
df.isnull().sum()
```

# *Univariate Analysis*

Univariate Analysis on  Numerical Variables

```python
# Now check for any NaN values that were created during the conversion
df['streams'].isna().sum()

# Attempt to convert problematic columns to numeric
df['streams'] = pd.to_numeric(df['streams'], errors='coerce')

a1 = df['streams'].mean()
a2 = df['streams'].median()
a3 = df['streams'].std()
print(a1)
print(a2)
print(a3)
```

```python
b1 = df['bpm'].mean()
b2 = df['bpm'].median()
b3 = df['bpm'].std()
print(b1)
print(b2)
print(b3)
```

Plotting Distribution of Numerical Variables

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.histplot(df['streams'], kde=True)
plt.title('Distribution of Numerical Column')
plt.show()

plt.tight_layout()
```

```python
# Box Plot
plt.figure(figsize=(8,6))
sns.boxplot(df['streams'])
plt.title('Box Plot of Numerical Column')
plt.show()
```

Categorical Data Analysis

```python
df['track_name'].value_counts()
sns.countplot(x='track_name', data=df)
plt.title('Category Distribution')
plt.show()
```

# *Multivariate Analysis*

Multivariate Analysis

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Correlation Heatmap
correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Scatter plot
sns.scatterplot(x='streams', y='bpm', data=df)
plt.title('Scatter Plot Between Popularity and Duration')
plt.show()
```

```python
# pair plot
sns.pairplot(df[['streams', 'bpm', 'track_name']])
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='track_name', y='streams', data=df)
plt.title('Boxplot of Numerical Column by Categories')
plt.show()
```
